```python
# Import the required libraries
from flask import Flask, request, jsonify, render_template, Response
import os
from pathlib import Path
import subprocess
import cv2

from flask_ngrok import run_with_ngrok
app = Flask(__name__)
run_with_ngrok(app)

@app.route('/')
def index():
    try:
        return render_template("index.html")
    except Exception as e:
        return jsonify({"error": str(e)})

# Define the route for video upload and processing
@app.route('/upload', methods=['POST'])
def upload_video():
    # Check if a video file was included in the request
    if 'video' not in request.files:
        return jsonify({"error": "No video file provided"})

    # Get the uploaded video file
    video_file = request.files['video']

    # Check if the file has an allowed extension (e.g., .mp4)
    if video_file.filename == '':
        return jsonify({"error": "Invalid video file"})  # Fixed typo

    # Save the video file to a temporary location
    video_path = "temp_video.mp4"
    video_file.save(video_path)

    # Define the YOLOv5 model and configuration files
    weights_path = "runs/train/fire_detection/weights/best.pt"
    config_path = "models/yolov5s.yaml"

    # Define the output directory for processed video and images
    output_dir = "output"

    # Create the output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # Use YOLOv5 to perform object detection on the uploaded video
    command = f"python detect.py --weights {weights_path} --source {video_path} --img-size 640 --conf-thres 0.25 --output {output_dir}"
    subprocess.run(command, shell=True, check=True)

    # Get the list of output image files
    image_files = [str(file) for file in Path(output_dir).glob("*.jpg")]

    # Define a generator function to serve processed video as chunks
    def generate():
        for image_path in image_files:
            img = cv2.imread(image_path)
            _, buffer = cv2.imencode('.jpg', img)
            yield (b'--frame\r\n'
                   b'Content-Type: image/jpeg\r\n\r\n' + buffer.tobytes() + b'\r\n')

    # Return the processed video as a response
    return Response(generate())
if __name__ == '__main__':
    app.run()
```
